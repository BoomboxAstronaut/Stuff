```python
modern = pd.read_csv(r'C:\Users\BBA\Coding\Resources\MiniDataSets\2020_States_Data')
old = pd.read_csv(r'C:\Users\BBA\Coding\Resources\MiniDataSets\1960_States_Data')
modern_arr = np.array(modern)
modern_arr = modern_arr.T[1:].T.astype(np.float32)
old_arr = np.array(old)
old_arr = old_arr.T[1:].T.astype(np.float32)
ostd = np.std(old_arr, 0)
mstd = np.std(modern_arr, 0)
omn = np.mean(old_arr, 0)
mmn = np.mean(modern_arr, 0)
modern_norm = np.asarray([[(y-mmn[i])/mstd[i] for y in x] for i, x in enumerate(modern_arr.T)]).T
old_norm = np.asarray([[(y-omn[i])/ostd[i] for y in x] for i, x in enumerate(old_arr.T)]).T
```

```python
fig, ax = plt.subplots()
im = ax.imshow(np.cov(modern_norm.T), cmap='PRGn')
labels = modern.columns[1:]

ax.set_xticks(np.arange(len(labels)), labels=labels)
ax.set_yticks(np.arange(len(labels)), labels=labels)
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

for i in range(5):
    for j in range(5):
        text = ax.text(j, i, round(np.cov(modern_norm.T)[i, j], 4), ha="center", va="center", color="k")

ax.grid(False)
ax.set_title("2020 Correlation Matrix")
fig.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots()
im = ax.imshow(np.cov(old_norm.T), cmap='PRGn')
labels = modern.columns[1:]

ax.set_xticks(np.arange(len(labels)), labels=labels)
ax.set_yticks(np.arange(len(labels)), labels=labels)
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

for i in range(5):
    for j in range(5):
        text = ax.text(j, i, round(np.cov(old_norm.T)[i, j], 4), ha="center", va="center", color="k")

ax.grid(False)
ax.set_title("1960 Correlation Matrix")
fig.tight_layout()
plt.show()
```
